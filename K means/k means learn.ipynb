```python
import pandas as pd

from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture as EM
from matplotlib import pyplot as plt

names=["Sepal_height", "Sepal_width", "Petal_height", "Petal_width", "Class"]
data = pd.read_csv("8dataset.csv", names=names)

X=data.iloc[:,0:-1]
label = {"Iris-setosa": 0, "Iris-versicolor": 1, "Iris-virginica": 2}
y = [label[c] for c in data.iloc[:,-1]]


def drawGraph(labels, title, n):
    plt.subplot(2,2,n)
    plt.scatter(X.Petal_height, X.Petal_width, c=labels)
    plt.title(title)
    plt.xlabel("Petal Height")
    plt.ylabel("Petal Width")
#     plt.show()
    pass

plt.figure(figsize=(10,10))
drawGraph(y, "Real Cluster", 1)

km = KMeans(3).fit(X)
drawGraph(km.labels_, "K Means Cluster", 2)

em = EM(3).fit(X)
drawGraph(em.predict(X), "EM Cluster", 3)
```
